```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
avg500_df = pd.read_csv('csv/avg_500.csv')
avg25000_df = pd.read_csv('csv/avg_25000.csv')

# Files
files20  = pd.read_csv('csv/files_20t20f.csv')
files100 = pd.read_csv('csv/files_100t100f.csv')
```

```python
sns.set_context("notebook")
```

```python
plt.figure(figsize=(10,5))
avg500_plot = sns.scatterplot(x='threads_maximo',y='time', data=avg500_df)
avg500_plot = avg500_plot.set(xticks=range(31))
plt.ylabel('time (ms)')
plt.xlabel('threads')
plt.title('Tiempo de acuerdo a threads (500 palabras)')
plt.savefig('maximo500.png')
```

```python
plt.figure(figsize=(10,5))
avg25000_plot = sns.scatterplot(x='threads_maximo',y='time', data=avg25000_df)
avg25000_plot = avg25000_plot.set(xticks=range(31))
plt.ylabel('time (ms)')
plt.xlabel('threads')
plt.title('Tiempo de acuerdo a threads (25000 palabras)')
plt.savefig('maximo25000.png')
```

```python
plt.figure(figsize=(8,5))
files20_plot = sns.scatterplot(x='threads_files',y='time', data=files20)
files20_plot = files20_plot.set(xticks=range(21))
plt.ylabel('time (ms)')
plt.xlabel('threads')
plt.title('Tiempo de acuerdo a threads (25000 palabras)')
plt.savefig('files20.png')
```
